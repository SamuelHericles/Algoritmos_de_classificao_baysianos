### 2.2) Classificadores Bayesianos
- Implemente os seguintes classificadores bayesianos:    

I) QDA (utilizando a formula√ß√£o completa do discriminante bayesiano).

```python
def qda(X,y,verbose=False):
    # 1- calcula a probilidade a priori[1x1]
    probs_p = fp.prob_a_priori(X)

    # 2- vetor m√©dio por classe[1x34]
    u = fp.vetor_medio_p_classe(X)

    
    # 3- matriz covariancia por classe[34x34]
    mcov = []
    for i in X['c35'].unique():
        mcov.append(fp.matriz_covariancia(X.query('c35=='+str(i))))

    # 4- Fazer a corre√ß√£o lamdba A = A + xI na matriz de covariancia
    mcov_lambda = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_lambda.append(fp.correcao_lambda(mcov[i]))
    
    # 5- Calcular a inversa da matriz de covair√¢ncia regulada lambda
    mcov_inv = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_inv.append(fp.matriz_inversa(mcov_lambda[i]))

    # 6- Calcular o determinante de cada matriz de covari√¢ncia regulada
    det = []
    for i in range(X['c35'].unique().shape[0]):
        det.append(fp.determinante_matriz_covariancia(mcov_lambda[i]))    
    
    # Teste, pegar agora um elementro da base de teste e aplicar
    # min: ln|det(mcov_classe_i)| + 
    #       (elem_teste-vetor_medio_classe_i).T*mcov_classe_i_inv*(elem_teste-vetor_md_classe_i)
    #       - 2*ln(P_priori(classe_i)) 

    y_pred = []
    for i in y.index.values:
        valores = []
        for j in range(X['c35'].unique().shape[0]):
            valores.append(fp.teste_elem(y.iloc[i,:-1],u.iloc[j],mcov_inv[j],det[j],probs_p[j]))    
        y_pred.append(valores.index(min(valores))+1)
    
    df_m_confusao = fp.matriz_confusao(y['c35'],y_pred)
    
    if verbose: fp.plot_matiz_de_confusao(df_m_confusao)
        
    return fp.parametros_de_medicao(df_m_confusao)
```

```python
X,y = fp.kfold_shuffle()
print(qda(X,y,verbose=True))
```

```python
qda_metricas = pd.DataFrame({})
for _ in range(5):
    X,y = fp.kfold_shuffle()
    qda_metricas = qda_metricas.append(qda(X,y,verbose=False),ignore_index=True)
print(f'M√©dia ds m√©tricas \n{qda_metricas.mean()}')
```

II) Naive Bayes (considerando que os atributos do problema s√£o descorrelacionados)

```python
def naives_bayes(X,y,verbose=False):
    # 1- calcula a probilidade a priori[1x1]
    probs_p = fp.prob_a_priori(X)

    # 2- vetor m√©dio por classe[1x34]
    u = fp.vetor_medio_p_classe(X)

    mcov = []
    for i in X['c35'].unique():
        mcov.append(fp.matriz_covariancia(X.query('c35=='+str(i))))
    
    # 3- matriz covariancia por classe[34x34] sabendo que s√£o descorrelacionados
    mcov_descorrelacionada = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_descorrelacionada.append(fp.limpar_covariancia(mcov[i]))

    # 4- Fazer a corre√ß√£o lamdba A = A + xI na matriz de covariancia
    mcov_lambda = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_lambda.append(fp.correcao_lambda(mcov_descorrelacionada[i]))

    # 5- Calcular a inversa da matriz de covair√¢ncia regulada lambda
    mcov_inv = []
    for i in range(X['c35'].unique().shape[0]):
        mcov_inv.append(fp.matriz_inversa(mcov_lambda[i]))
    
    # 6- Calcular o determinante de cada matriz de covari√¢ncia regulada
    det = []
    for i in range(X['c35'].unique().shape[0]):
        det.append(fp.determinante_matriz_covariancia(mcov_lambda[i]))    
    
    # Teste, pegar agora um elementro da base de teste e aplicar
    # min: ln|det(mcov_classe_i)| + 
    #       (elem_teste-vetor_medio_classe_i).T*mcov_classe_i_inv*(elem_teste-vetor_md_classe_i)
    #       - 2*ln(P_priori(classe_i)) 
    
    y_pred = []
    for i in y.index.values:
        valores = []
        for j in range(X['c35'].unique().shape[0]):
            valores.append(fp.teste_elem(y.iloc[i,:-1],u.iloc[j],mcov_inv[j],det[j],probs_p[j]))    
        y_pred.append(valores.index(min(valores))+1)
        
    df_m_confusao = fp.matriz_confusao(y['c35'],y_pred)
    if verbose: fp.plot_matiz_de_confusao(df_m_confusao)
    
    return fp.parametros_de_medicao(df_m_confusao)
```

```python
X,y = fp.kfold_shuffle()
print(naives_bayes(X,y,verbose=True))
```

```python
naive_bayes_metricas = pd.DataFrame({})
for _ in range(5):
    X,y = fp.kfold_shuffle()
    naive_bayes_metricas = naive_bayes_metricas.append(naives_bayes(X,y,verbose=False),ignore_index=True)
print(f'M√©dia m√©tricas \n{naive_bayes_metricas.mean()}')
```

III) LDA (considerando que todos os classificadores possuem a mesma matriz de covari√¢ncia e a mesma probabilidade √† priori).

‚Ä¢ OBS: para calcular a matriz de covari√¢ncia agregada, deve-se calcular as matrizes de covari√¢ncia de cada classe, e calcular a m√©dia destas.

- Para cada classificador, utilizando o banco de dados ‚Äúdermatologia‚Äù realize o seguinte experimento:

    + Utilize a valida√ß√£o cruzada k-fold (k = 5) para gerar os resultados.
    + Para cada um dos 5 particionamentos dos dados (entre treinamento e teste) realize o seguinte procedimento:
        + A partir dos dados de treinamento, calcule a m√©dia e o desvio padr√£o de cada atributo e fa√ßa a normaliza√ß√£o z-score dos dados (tanto dos dados de treinamento como dos dados de teste).
        + Calcule as estat√≠sticas necess√°rias (matriz de covari√¢ncia, vetor m√©dio, probabilidade a priori...), para o classificador, a partir dos dados de treinamento.
        + Classifique os dados de teste.

```python
def lda(X,y,verbose=False):
    
    # 1- vetor m√©dio por classe[1x34]
    u= []
    for i in range(6):
        u.append(fp.vetor_medio(X.query('c35=='+str(i+1))))
    
    # 2- mcov_medio ou agregaddo √© [34x34] somar todas as mcov de n classes e fazer a m√©dia mcov/n classes [34x34]
    mcovs = 0
    for i in X['c35'].unique():
        mcovs  += fp.matriz_covariancia(X.query('c35=='+str(i)))
    mcov_medio = mcovs/6
    
    # 3- Fazer a corre√ß√£o lamdba A = A + xI na matriz de covariancia
    mcov_medio_lambda = fp.correcao_lambda(mcov_medio)
 
    # 5- Calcular a inversa da matriz de covair√¢ncia regulada lambda
    mcov_medio_lambda_inv_d = fp.matriz_inversa(mcov_medio_lambda)
    
    # 4- Teste, pegar agora um elementro da base de teste e aplicar
    # min: (elem_teste-vetor_medio_classe_i).T*mcov_classe_i_inv_media(agredada)*(elem_teste-vetor_md_classe_i)
    y_pred = []
    for i in y.index.values:
        valores = []
        for j in range(len(u)):
            valores.append(fp.teste_elem_lda(y.iloc[i,:-1],u[j],mcov_medio_lambda_inv_d))    
        y_pred.append(valores.index(min(valores))+1)
        
    df_m_confusao = fp.matriz_confusao(y['c35'],y_pred)
    if verbose: fp.plot_matiz_de_confusao(df_m_confusao)
    
    return fp.parametros_de_medicao(df_m_confusao)
```

```python
X,y = fp.kfold_shuffle()
print(lda(X,y,verbose=True))
```

```python
lda_metricas = pd.DataFrame({})
for _ in range(5):
    X,y = fp.kfold_shuffle()
    lda_metricas = lda_metricas.append(lda(X,y,verbose=False),ignore_index=True)
print(f'M√©dia m√©tricas \n{lda_metricas.mean()}')
```

2.3) LDA (CDA) como transforma√ß√£o linear.
- A partir do banco de dados dermatologia, realize o seguinte experimento:
    + Separe os dados entre treinamento (70%) e teste (30%).
    + A partir dos dados de treinamento, calcule a m√©dia e o desvio padr√£o de cada atributo e fa√ßa a normaliza√ß√£o z-score dos dados (tanto dos dados de treinamento como dos dados de teste).
    + A partir dos dados de treinamento, aplique o m√©todo LDA utilizando a abordagem de m√∫ltiplas proje√ß√µes (tamb√©m conhecido como Canonical Discriminant Analysis, CDA).
    + A partir dos autovalores calculados, defina uma quantidade de atributos necess√°ria para realizar a proje√ß√£o dos dados. (Obs: quantidade m√°xima = ‚Äún√∫mero de classes ‚Äì 1‚Äù)
    + Defina a matriz de proje√ß√£o W.
    + Transforme os dados de treinamento a partir da matriz W.
    + Utilize os dados transformados para treinar um classificador bayesiano (√† sua escolha)
    + Utilize a matriz de proje√ß√£o transformar os dados de teste.
    + Classifique os dados de teste e obtenha a taxa de acerto do classificador.

```python
def CDA(X):    

    # 1 - Calcular a matriz Sw
    Sw = 0
    for i in range(6):
        Sw += fp.matriz_covariancia(X.query('c35=='+str(i+1)))
        
    # 2 - Calcular o vetor m√©dio de cada classse
    us = fp.vetor_medio_p_classe(X)

    # 3 - Calcular a media dos vetores m√©dios
    u = us.mean()
    
    # 4 - Calcular a matriz de proje√ß√£o Sb
    Sb = 0
    for i in X['c35'].unique():
        Sb  += fp.calcula_projecao(X.query('c35=='+str(i)),u,us.iloc[i-1]).values
    
    # 5 - Calcular os autovalores e autovetores
    m_aa_av = fp.matriz_inversa(Sw)*Sb
    
    a_valores,a_vetores = np.linalg.eig(m_aa_av)
    df_a_valores = pd.DataFrame(a_valores)
    df_a_vetores = pd.DataFrame(a_vetores)
    
    # 6 - Procurar os autovetores dos maiores autovalores
    W = pd.DataFrame({})
    for i in df_a_valores.sort_values(by=0,ascending=True)[:5].index.values:
        W = W.append(df_a_vetores.iloc[i],ignore_index=True)

    return W
```

```python
naives_bayes_metricas = pd.DataFrame({})
for _ in range(5):
    X,y = fp.treino_teste_70_30(derma_dados)
    W = CDA(X)
    Xtr_t =  fp.transformar_dados(W,X)
    Xts_t =  fp.transformar_dados(W,y)
    naives_bayes_metricas = naives_bayes_metricas.append(naives_bayes(Xtr_t,Xts_t,verbose=False),ignore_index=True)
print(f'M√©dia m√©tricas \n{naive_bayes_metricas.mean()}')
```

```python
qda_metricas = pd.DataFrame({})
for _ in range(5):
    X,y = fp.treino_teste_70_30(derma_dados)
    W = CDA(X)
    Xtr_t =  fp.transformar_dados(W,X)
    Xts_t =  fp.transformar_dados(W,y)
    qda_metricas = qda_metricas.append(qda(Xtr_t,Xts_t,verbose=False),ignore_index=True)
print(f'M√©dia ds m√©tricas \n{qda_metricas.mean()}')
```
