# Probabilidade condicional

```python
def prob_cond(A,B,df=derma_dados):
    PA = A.shape[0]/df.shape[0]
    PB = B.shape[0]/df.shape[0]
    PAB = PA*PB
    return (PAB*PA)/PB
```

```python
def dist_normal(base):
    parte_1 = 1/np.sqrt(2*math.pi*base.var())
    expoente = -pow(base-base.mean(),2)/2*(pow(base.var(),2))
    parte_2 = pow(math.e, expoente)
    return parte_1*parte_2
def dist_normal_multivariada(base):
    parte_1 = (1/np.sqrt(2*math.pi*sum(base)))
    parte_2 = base - base.mean()
    parte_2 = -(1/2)*(parte_2.transpose())*(sum(parte_2))
    return (parte_1*pow(math.e,parte_2))
```

# Vetor médio

```python
def vetor_medio(base):
    vetor_medio = []
    for i,j in zip(base.columns[:-1],[i for i in range(base.shape[1])]):
        vetor_medio.append(base[str(i)].mean())
    df = pd.DataFrame(data=vetor_medio,index=base.columns[:-1])
    return df.T
vt_medio = vetor_medio(derma_dados)
vt_medio
```

# Vetor de variâncias

```python
def vetor_variancias(base):
    vetor_medio = []
    for i,j in zip(base.columns[:-1],[i for i in range(base.shape[1])]):
        vetor_medio.append(base[str(i)].var())
    df = pd.DataFrame(data=vetor_medio,index=base.columns[:-1])
    return df.T
vetor_variancias(derma_dados)
```

# Matriz de covariância

```python
def matriz_covariancia(base):
    df = pd.DataFrame(index=[base.columns[i] for i in range(base.shape[1])],columns=base.columns[:-1])
    for i,count_i in zip(base.columns[:-1],[i for i in range(base.shape[1])]):
        for j,count_j in zip(base.columns[:-1],[i for i in range(base.shape[1])]):
            df.iloc[count_i,count_j] = (sum(base[str(i)]*base[str(j)])#sum(xy)                                        
                                     - (sum(base[str(i)])*sum(base[str(j)]))#sum(x)*sum(y)
                                        /base.shape[0])/base.shape[0] #1/m 1/m
    return df.iloc[:-1]
```

```python
def norm_z_score(base_trenio, base_teste):
    media         = base_trenio[base_trenio.columns[:-1]].mean()
    desvio_padrao = base_trenio[base_trenio.columns[:-1]].std()
    base_trenio[base_trenio.columns[:-1]] = base_trenio[base_trenio.columns[:-1]]-media/desvio_padrao
    base_teste[base_teste.columns[:-1]]   = base_teste[base_teste.columns[:-1]]-media/desvio_padrao
    return base_trenio,base_teste
```

```python
def kfold_shuffle(base,k):
    X = pd.DataFrame({})
    y = pd.DataFrame({})
    
    index_teste_classe = sorted(np.random.choice(base.index.values,round((base.shape[0]/5))))
    X_classe = base.iloc[[i for i in base.index if i not in index_teste_classe],:]
    y_classe = base.iloc[index_teste_classe,:]

    X = X.append(X_classe,ignore_index=True)
    y = y.append(y_classe,ignore_index=True)

    return norm_z_score(X,y)
```

```python
def prob_a_priori(base):
    probs_p = []
    for i in base['c35'].unique():
        probs_p.append(base.query('c35=='+str(i))['c35'].shape[0]/base.shape[0])
    return probs_p
```

```python
def teste_elem(X,X_medio,mcov_inv,determinante,probabilidade):
    det_log  = np.log(abs(determinante))
    m_linha  = (X-X_medio.T).values
    m_coluna = (X-X_medio.T).T.values
    log_prob = np.log(probabilidade)
    resultado = det_log + (m_linha @ mcov_inv.values) @ (m_coluna) - 2*log_prob
    return resultado[0][0]
```

```python
def get_acc(y_pred,y_true):
    return round(sum(y_pred==y_true)/y_true.shape[0],4)
```

```python
def teste_elem_lda(X,X_medio,mcov_inv):
    m_linha   = (X-X_medio.T).values
    m_coluna  = (X-X_medio.T).T.values
    resultado = (m_linha @ mcov_inv) @ m_coluna
    return resultado[0][0]
```

```python
def limpar_covariancia(matriz_covariancia):
    for i in range(34):
        for j in range(34):
            if i!=j:
                matriz_covariancia.iloc[i,j]=0
    return matriz_covariancia
```

# I) QDA

```python
# Antes de tudos normalizar z-score os dados com as info do dados de treino(mean e std)
# iSSO TUDO PARA 6 CLASSES

def qda(X,y):
    # 1- calcula a probilidade a priori[1x1]
    probs_p = prob_a_priori(X)

    # 2- vetor médio por classe[1x34]
    vt_medio_1 = vetor_medio(X.query('c35==1'))
    vt_medio_2 = vetor_medio(X.query('c35==2'))
    vt_medio_3 = vetor_medio(X.query('c35==3'))
    vt_medio_4 = vetor_medio(X.query('c35==4'))
    vt_medio_5 = vetor_medio(X.query('c35==5'))
    vt_medio_6 = vetor_medio(X.query('c35==6'))


    # 3- matriz covariancia por classe[34x34]
    mcov_1 = matriz_covariancia(X.query('c35==1'))
    mcov_2 = matriz_covariancia(X.query('c35==2'))
    mcov_3 = matriz_covariancia(X.query('c35==3'))
    mcov_4 = matriz_covariancia(X.query('c35==4'))
    mcov_5 = matriz_covariancia(X.query('c35==5'))
    mcov_6 = matriz_covariancia(X.query('c35==6'))

    # 4- Fazer a correção lamdba A = A + xI na matriz de covariancia
    mcov_1_lambda = mcov_1 + np.identity(mcov_1.shape[0], dtype=float)*0.01
    mcov_2_lambda = mcov_2 + np.identity(mcov_2.shape[0], dtype=float)*0.01
    mcov_3_lambda = mcov_3 + np.identity(mcov_3.shape[0], dtype=float)*0.01
    mcov_4_lambda = mcov_4 + np.identity(mcov_4.shape[0], dtype=float)*0.01
    mcov_5_lambda = mcov_5 + np.identity(mcov_5.shape[0], dtype=float)*0.01
    mcov_6_lambda = mcov_6 + np.identity(mcov_6.shape[0], dtype=float)*0.01

    # 5- Calcular a inversa da matriz de covairância regulada lambda
    mcov_1_lambda_inv = pd.DataFrame(np.linalg.inv(np.matrix(mcov_1_lambda.values, dtype='float')))
    mcov_2_lambda_inv = pd.DataFrame(np.linalg.inv(np.matrix(mcov_2_lambda.values, dtype='float')))
    mcov_3_lambda_inv = pd.DataFrame(np.linalg.inv(np.matrix(mcov_3_lambda.values, dtype='float')))
    mcov_4_lambda_inv = pd.DataFrame(np.linalg.inv(np.matrix(mcov_4_lambda.values, dtype='float')))
    mcov_5_lambda_inv = pd.DataFrame(np.linalg.inv(np.matrix(mcov_5_lambda.values, dtype='float')))
    mcov_6_lambda_inv = pd.DataFrame(np.linalg.inv(np.matrix(mcov_6_lambda.values, dtype='float')))

    # 6- Calcular o determinante de cada matriz de covariância regulada
    det_1 = np.linalg.det(np.matrix(mcov_1_lambda.values, dtype='float'))
    det_2 = np.linalg.det(np.matrix(mcov_2_lambda.values, dtype='float'))
    det_3 = np.linalg.det(np.matrix(mcov_3_lambda.values, dtype='float'))
    det_4 = np.linalg.det(np.matrix(mcov_4_lambda.values, dtype='float'))
    det_5 = np.linalg.det(np.matrix(mcov_5_lambda.values, dtype='float'))
    det_6 = np.linalg.det(np.matrix(mcov_6_lambda.values, dtype='float'))


    # Teste, pegar agora um elementro da base de teste e aplicar
    # min: ln|det(mcov_classe_i)| + 
    #       (elem_teste-vetor_medio_classe_i).T*mcov_classe_i_inv*(elem_teste-vetor_md_classe_i)
    #       - 2*ln(P_priori(classe_i)) 

    # elem_teste é X da questão, literalmente
    # o que o min retorna?
    # min retorna um valor 1,2,3,4,5,6 sendo este a classe

    aux = []
    for i in y.index.values:
        valores = []
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_1,mcov_1_lambda_inv,det_1,probs_p[0]))    
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_2,mcov_2_lambda_inv,det_2,probs_p[1]))
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_3,mcov_3_lambda_inv,det_3,probs_p[2]))    
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_4,mcov_4_lambda_inv,det_4,probs_p[3]))
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_5,mcov_5_lambda_inv,det_5,probs_p[4]))
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_6,mcov_6_lambda_inv,det_6,probs_p[5]))
        aux.append(valores.index(min(valores))+1)
    print('Acrurácia de ',get_acc(aux,y['c35'])*100,'%')
    return get_acc(aux,y['c35'])
```

```python
# Antes de tudos normalizar z-score os dados com as info do dados de treino(mean e std)
# iSSO TUDO PARA 6 CLASSES
# Antes de tudo diagonalizar a mcov 

def nayves_bayes(X,y):
    # 1- calcula a probilidade a priori[1x1]
    probs_p = prob_a_priori(X)

    # 2- vetor médio por classe[1x34]
    vt_medio_1 = vetor_medio(X.query('c35==1'))
    vt_medio_2 = vetor_medio(X.query('c35==2'))
    vt_medio_3 = vetor_medio(X.query('c35==3'))
    vt_medio_4 = vetor_medio(X.query('c35==4'))
    vt_medio_5 = vetor_medio(X.query('c35==5'))
    vt_medio_6 = vetor_medio(X.query('c35==6'))


    # 3- matriz covariancia por classe[34x34] sabendo que são descorrelacionados
    mcov_1_descorrelacionada = limpar_covariancia(mcov_1)
    mcov_2_descorrelacionada = limpar_covariancia(mcov_2)
    mcov_3_descorrelacionada = limpar_covariancia(mcov_3)
    mcov_4_descorrelacionada = limpar_covariancia(mcov_4)
    mcov_5_descorrelacionada = limpar_covariancia(mcov_5)
    mcov_6_descorrelacionada = limpar_covariancia(mcov_6)

    # 4- Fazer a correção lamdba A = A + xI na matriz de covariancia
    mcov_1_lambda_d = mcov_1_descorrelacionada + np.identity(mcov_1_descorrelacionada.shape[0], dtype=float)*0.01
    mcov_2_lambda_d = mcov_1_descorrelacionada + np.identity(mcov_1_descorrelacionada.shape[0], dtype=float)*0.01
    mcov_3_lambda_d = mcov_1_descorrelacionada + np.identity(mcov_1_descorrelacionada.shape[0], dtype=float)*0.01
    mcov_4_lambda_d = mcov_1_descorrelacionada + np.identity(mcov_1_descorrelacionada.shape[0], dtype=float)*0.01
    mcov_5_lambda_d = mcov_1_descorrelacionada + np.identity(mcov_1_descorrelacionada.shape[0], dtype=float)*0.01
    mcov_6_lambda_d = mcov_1_descorrelacionada + np.identity(mcov_1_descorrelacionada.shape[0], dtype=float)*0.01

    # 5- Calcular a inversa da matriz de covairância regulada lambda
    mcov_1_lambda_inv_d = pd.DataFrame(np.linalg.inv(np.matrix(mcov_1_lambda_d.values, dtype='float')))
    mcov_2_lambda_inv_d = pd.DataFrame(np.linalg.inv(np.matrix(mcov_2_lambda_d.values, dtype='float')))
    mcov_3_lambda_inv_d = pd.DataFrame(np.linalg.inv(np.matrix(mcov_3_lambda_d.values, dtype='float')))
    mcov_4_lambda_inv_d = pd.DataFrame(np.linalg.inv(np.matrix(mcov_4_lambda_d.values, dtype='float')))
    mcov_5_lambda_inv_d = pd.DataFrame(np.linalg.inv(np.matrix(mcov_5_lambda_d.values, dtype='float')))
    mcov_6_lambda_inv_d = pd.DataFrame(np.linalg.inv(np.matrix(mcov_6_lambda_d.values, dtype='float')))

    # 5- Teste, pegar agora um elementro da base de teste e aplicar
    # min: ln|det(mcov_classe_i)| + 
    #       (elem_teste-vetor_medio_classe_i).T*mcov_classe_i_inv*(elem_teste-vetor_md_classe_i)
    #       - 2*ln(P_priori(classe_i)) 
    aux = []
    for i in y.index.values:
        valores = []
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_1,mcov_1_lambda_inv_d,det_1,probs_p[0]))    
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_2,mcov_2_lambda_inv_d,det_2,probs_p[1]))
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_3,mcov_3_lambda_inv_d,det_3,probs_p[2]))    
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_4,mcov_4_lambda_inv_d,det_4,probs_p[3]))
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_5,mcov_5_lambda_inv_d,det_5,probs_p[4]))
        valores.append(teste_elem(y.iloc[i,:-1],vt_medio_6,mcov_6_lambda_inv_d,det_6,probs_p[5]))
        aux.append(valores.index(min(valores))+1)
    print('Acrurácia de ',get_acc(aux,y['c35'])*100,'%')
    return get_acc(aux,y['c35'])
```

# III)LDA

```python
# Considere que a prob a priori eh igual para todos, então não se usa aqui
# Considere que a matriz de covariancia é igual par atodas as classes, então irei calcular a media mcov das n classes
# Antes de tudos normalizar z-score os dados com as info do dados de treino(mean e std)
# iSSO TUDO PARA 6 CLASSES

def lda(X,y):
    # 1- vetor médio por classe[1x34]
    vt_medio_1 = vetor_medio(X.query('c35==1'))
    vt_medio_2 = vetor_medio(X.query('c35==2'))
    vt_medio_3 = vetor_medio(X.query('c35==3'))
    vt_medio_4 = vetor_medio(X.query('c35==4'))
    vt_medio_5 = vetor_medio(X.query('c35==5'))
    vt_medio_6 = vetor_medio(X.query('c35==6'))


    # 2- mcov_medio ou agregaddo é [34x34] somar todas as mcov de n classes e fazer a média mcov/n classes [34x34]
    mcov_1 = matriz_covariancia(X.query('c35==1'))
    mcov_2 = matriz_covariancia(X.query('c35==2'))
    mcov_3 = matriz_covariancia(X.query('c35==3'))
    mcov_4 = matriz_covariancia(X.query('c35==4'))
    mcov_5 = matriz_covariancia(X.query('c35==5'))
    mcov_6 = matriz_covariancia(X.query('c35==6'))
    mcov_medio = mcov_1+mcov_2+mcov_3+mcov_4+mcov_5+mcov_6
    mcov_medio = mcov_medio/6

    # 3- Fazer a correção lamdba A = A + xI na matriz de covariancia
    mcov_medio_lambda = mcov_medio + np.identity(mcov_medio.shape[0], dtype=float)*0.01

    # 5- Calcular a inversa da matriz de covairância regulada lambda
    mcov_medio_lambda_inv_d = pd.DataFrame(np.linalg.inv(np.matrix(mcov_medio_lambda.values, dtype='float')))
    
    # 4- Teste, pegar agora um elementro da base de teste e aplicar
    # min: (elem_teste-vetor_medio_classe_i).T*mcov_classe_i_inv_media(agredada)*(elem_teste-vetor_md_classe_i)
    aux = []
    for i in y.index.values:
        valores = []
        valores.append(teste_elem_lda(y.iloc[i,:-1],vt_medio_1,mcov_medio_lambda_inv_d))    
        valores.append(teste_elem_lda(y.iloc[i,:-1],vt_medio_2,mcov_medio_lambda_inv_d))
        valores.append(teste_elem_lda(y.iloc[i,:-1],vt_medio_3,mcov_medio_lambda_inv_d))    
        valores.append(teste_elem_lda(y.iloc[i,:-1],vt_medio_4,mcov_medio_lambda_inv_d))
        valores.append(teste_elem_lda(y.iloc[i,:-1],vt_medio_5,mcov_medio_lambda_inv_d))
        valores.append(teste_elem_lda(y.iloc[i,:-1],vt_medio_6,mcov_medio_lambda_inv_d))
        aux.append(valores.index(min(valores))+1)
    print('Acrurácia de ',get_acc(aux,y['c35'])*100,'%')
    return get_acc(aux,y['c35'])
```
